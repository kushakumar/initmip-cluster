```python
import os
import shutil

# Source directory
source_dir = "/projects/grid/ghub/ISMIP6/initMIP/GrIS/output_regridded/"

# Destination directory
dest_dir = "/user/kushakum/initmip-cluster/data/models/"

# Function to copy files while preserving directory structure
def copy_files_preserving_structure(src_dir, dst_dir):
    for root, dirs, files in os.walk(src_dir):
        # Construct the destination directory path
        relative_path = os.path.relpath(root, src_dir)
        dest_path = os.path.join(dst_dir, relative_path)

        # Ensure the destination directory exists
        os.makedirs(dest_path, exist_ok=True)

        # Copy each file
        for file in files:
            if file.endswith(".nc"):
                src_file = os.path.join(root, file)
                dst_file = os.path.join(dest_path, file)
                shutil.copy(src_file, dst_file)

# Copy files
copy_files_preserving_structure(source_dir, dest_dir)

print("Files have been copied successfully with the directory structure preserved.")
```
